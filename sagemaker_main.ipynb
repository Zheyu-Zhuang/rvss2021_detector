```python
import os
import json
import boto3
import sagemaker
from sagemaker.pytorch import PyTorch
```

```python
sagemaker_session = sagemaker.Session()

# The bucket containig our input data
source_bucket = "s3://rvssdata"
output = "s3://rvssdata/weights/"

role = sagemaker.get_execution_role()

estimator = PyTorch(entry_point = "main.py",
                    source_dir='scripts',
                    role = role,
                    instance_type = "ml.p2.xlarge",
                    instance_count = 1,
                    framework_version='1.6.0',
                    py_version='py3',
                    output_path = output,
                    hyperparameters = {'epoch': 20,
                                       'batch_size': 64,
                                       'lr': 5e-4,
                                       'scheduler_gamma' : 0.1,
                                       'scheduler_step' : 10
                                      }
                   )

estimator.fit({'train': source_bucket + '/fruit_1k'})
```
